```python
# solution 
from karel.stanfordkarel import *

"""
Karel should fill the whole world with beepers.
"""

def main():
    while front_is_clear():
        fill_row()
        move_up()

def fill_row():
    while front_is_clear():
        put_beeper()
        move()
    put_beeper()
    turn_around()
    while front_is_clear():
        move()
    turn_around()

def move_up():
    turn_left()
    if front_is_clear():
        move()
        turn_right()
    else:
        # this is to handle the final row scenario
        turn_right()
        move()

def turn_right():
    for i in range(3):
        turn_left()

def turn_around():
    for i in range(2):
        turn_left()

# There is no need to edit code beyond this point
if __name__ == '__main__':
    main()
```
