```python
import tkinter as tk
import random

# Function to move the ball
def move_ball():
    global dx, dy
    canvas.move(ball, dx, dy)
    x1, y1, x2, y2 = canvas.coords(ball)
    if x1 <= 0 or x2 >= WIDTH:
        dx *= -1
        change_ball_color()
    if y1 <= 0 or y2 >= HEIGHT:
        dy *= -1
        change_ball_color()
    root.after(DELAY, move_ball)

# Function to change the ball color
def change_ball_color():
    colors = ["red", "orange", "yellow", "green", "blue", "purple"]
    canvas.itemconfig(ball, fill=random.choice(colors))

# Constants
WIDTH = 600
HEIGHT = 400
DELAY = 10
BALL_SIZE = 30

# Initial ball direction
dx = 2
dy = 2

# Create the main window
root = tk.Tk()
root.title("Bouncing Ball Game")

# Create the canvas
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="black")
canvas.pack()

# Create the ball
ball = canvas.create_oval(WIDTH/2 - BALL_SIZE/2, HEIGHT/2 - BALL_SIZE/2,
                          WIDTH/2 + BALL_SIZE/2, HEIGHT/2 + BALL_SIZE/2,
                          fill="white")

# Start moving the ball
move_ball()

# Run the main loop
root.mainloop()
```

```python
# snake gameimport pygame
import random

# Initialize Pygame
pygame.init()

# Set up the game window
WIDTH, HEIGHT = 800, 600
window = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Snake Game")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

# Set up the snake and food
snake_pos = [100, 50]
snake_body = [[100, 50], [90, 50], [80, 50]]
food_pos = [random.randrange(1, (WIDTH//10)) * 10, random.randrange(1, (HEIGHT//10)) * 10]
food_spawn = True

# Set up game variables
direction = 'RIGHT'
change_to = direction
score = 0

# Set up game clock
clock = pygame.time.Clock()

# Main game loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                change_to = 'UP'
            elif event.key == pygame.K_DOWN:
                change_to = 'DOWN'
            elif event.key == pygame.K_LEFT:
                change_to = 'LEFT'
            elif event.key == pygame.K_RIGHT:
                change_to = 'RIGHT'

    # Validate the direction
    if change_to == 'UP' and direction != 'DOWN':
        direction = 'UP'
    if change_to == 'DOWN' and direction != 'UP':
        direction = 'DOWN'
    if change_to == 'LEFT' and direction != 'RIGHT':
        direction = 'LEFT'
    if change_to == 'RIGHT' and direction != 'LEFT':
        direction = 'RIGHT'

    # Update snake position
    if direction == 'UP':
        snake_pos[1] -= 10
    if direction == 'DOWN':
        snake_pos[1] += 10
    if direction == 'LEFT':
        snake_pos[0] -= 10
    if direction == 'RIGHT':
        snake_pos[0] += 10

    # Snake body mechanism
    snake_body.insert(0, list(snake_pos))
    if snake_pos[0] == food_pos[0] and snake_pos[1] == food_pos[1]:
        score += 1
        food_spawn = False
    else:
        snake_body.pop()

    if not food_spawn:
        food_pos = [random.randrange(1, (WIDTH//10)) * 10, random.randrange(1, (HEIGHT//10)) * 10]
    food_spawn = True

    # Graphics
    window.fill(BLACK)
    for pos in snake_body:
        pygame.draw.rect(window, GREEN, pygame.Rect(pos[0], pos[1], 10, 10))
    pygame.draw.rect(window, RED, pygame.Rect(food_pos[0], food_pos[1], 10, 10))

    # Game Over conditions
    if snake_pos[0] < 0 or snake_pos[0] > WIDTH-10:
        pygame.quit()
        quit()
    if snake_pos[1] < 0 or snake_pos[1] > HEIGHT-10:
        pygame.quit()
        quit()
    for block in snake_body[1:]:
        if snake_pos[0] == block[0] and snake_pos[1] == block[1]:
            pygame.quit()
            quit()

    # Display score
    font = pygame.font.SysFont(None, 25)
    score_text = font.render("Score: " + str(score), True, WHITE)
    window.blit(score_text, [10, 10])

    # Refresh game screen
    pygame.display.update()
    # Refresh rate
    clock.tick(20)

```
